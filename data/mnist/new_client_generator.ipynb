```python
all_data = []
for user in data['user_data'].keys():
    feature = np.array(data['user_data'][user]['x'])
    label = np.array(data['user_data'][user]['y']).reshape(-1, 1)
    c = np.concatenate((feature, label), axis=1)
    all_data.extend(c)
    
all_data = np.array(all_data)
all_data.shape
```

```python
np.random.shuffle(all_data)

fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(14, 7), sharey=True)
fig.subplots_adjust(hspace=0.5)

for i in range(20):
    x = all_data[i][:-1]
    y = all_data[i][-1]
    axes[i//5, i%5].imshow(x.reshape(28,28))
    axes[i//5, i%5].set_title(y)
```
