```python
import json
import re
import plotly.express
import plotly.io as pio
pio.kaleido.scope.mathjax = None # prevent using MathJax
```

```python
def read_from_html(file_name):
    with open(file_name) as f:
        html = f.read()
    call_arg_str = re.findall(r'Plotly\.newPlot\((.*)\)', html[-2**16:])[0]
    call_args = json.loads(f'[{call_arg_str}]')
    plotly_json = {'data': call_args[1], 'layout': call_args[2]}    
    return plotly.io.from_json(json.dumps(plotly_json))
```

```python
import os

list_file = []
def get_all_file(dir_or_file):
    if os.path.isfile(dir_or_file):
        list_file.append(dir_or_file)
    else:
        list_file_folder = os.listdir(dir_or_file)
        for item in list_file_folder:
            if item in ['bla.py', '.ipynb_checkpoints', 'json', 'format image.ipynb']:
                continue
            dir_ = os.path.join(dir_or_file, item)
            get_all_file(dir_)

get_all_file('./')
len(list_file)
```

```python
for item in list_file:
    print(item, end='\n\n')
```

```python
for idx, item in enumerate(list_file):
    if idx != 2 and idx != 3:
        os.remove(item)
```

```python
count = 0
for item in list_file:
    file_name = item.replace('.html', '.png')
    try:
        fig = read_from_html(item)
        fig.write_image(file_name)
    except:
        count += 1
        print(f'{count}. {file_name}')
```
